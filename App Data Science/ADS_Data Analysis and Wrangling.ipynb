## Types of Joins/Merge
1. inner join - intersection of keys (by default)
2. left join - use keys from left object
3. right join - use keys from right object
4. outer join - use union of keys

```python
# join using sub id
third=pd.merge(first,second,on='Sub_id',how='left')
third
```

```python
# join using sub id
third=pd.merge(first,second,on='Sub_id',how='right')
third
```

```python
# join using sub id
third=pd.merge(first,second,on='Sub_id',how='outer')
third
```

```python
# join using sub id
third=pd.merge(first,second,on=['id','Sub_id'])
third
```

```python
# join using sub id
third=pd.merge(first,second,on=['id','Sub_id'], how='left')
third
```

# Handle duplicate values

```python
data=pd.DataFrame({'A':[1,2,2,2,4],
                  'B':[5,6,6,6,8]})
```

```python
data
```

```python
#identify duplicate rows
dup_rows=data[data.duplicated()]
dup_rows
```

```python
data.duplicated()
```

```python
data[data.duplicated()].shape[0]
```

```python
# case-1 drop all duplicate rows
df=data.drop_duplicates()
print('Dataframe after removing duplicate')
df
```

```python
dup_rows=data[data.duplicated()==False]
dup_rows
```

```python
# case-2 keep first duplicate
df=data.drop_duplicates(keep='first')
print('Dataframe after removing duplicate')
df
```

```python
# case-2 keep last duplicate
df=data.drop_duplicates(keep='last')
print('Dataframe after removing duplicate')
df
```
