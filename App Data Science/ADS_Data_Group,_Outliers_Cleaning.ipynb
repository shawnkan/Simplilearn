# Outlier Analysis

## 1.5 IQR Rule

Low = Q1 - 1.5 * IQR

High = Q3 + 1.5 *IQR

```python
data.boxplot()
```

```python
data.boxplot(column=['MILES*'])
```

```python
#calculate IQR
IQR = data['MILES*'].quantile(0.75) - data['MILES*'].quantile(0.25)
IQR
```

```python
lower=data['MILES*'].quantile(0.25)-(1.5*IQR)
high=data['MILES*'].quantile(0.75)+(1.5*IQR)

print(lower)
print(high)
```

```python
# find out how many values more than high data
data[data['MILES*']>high]
```

```python
# find out how many values less than low data
data[data['MILES*']<lower]
```
