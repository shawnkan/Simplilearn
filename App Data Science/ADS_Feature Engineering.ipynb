# Scale the data

0 to 1

z-score / standard scaler
measure of how many std a particular data point is from mean of group of data

z= value-mean / std

```python
df_z_scaled=data.copy()
```

```python
#apply normalization technique (Z Score normalization)

for column in df_z_scaled:
  if df_z_scaled[column].dtypes == float:
    df_z_scaled[column]=(df_z_scaled[column]-df_z_scaled[column].mean())/df_z_scaled[column].std()
```

```python
df_z_scaled
```

```python
# min max scaling
# it scales the values of a feature to specific range between 0 and 1
# Xscaled = (X-Xmin)/(Xmax-Xmin)
df_min_max_scaled=data.copy()
```

```python
for column in df_min_max_scaled.columns:
    if df_min_max_scaled[column].dtypes==float:
      df_min_max_scaled[column]=(df_min_max_scaled[column]-df_min_max_scaled[column].min())/(df_min_max_scaled[column].max()-df_min_max_scaled[column].min())
```

```python
# sklearn one hot encoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
#'encoder' can be anyting
#[0] is apply on first column
# pass through is columns not affected passthrough
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')
X=np.array(ct.fit_transform(X))
```
