# Important for machine learning

```python
# find out variable which are highly coorelated with sale price
k=10
cols=house_data.corr().nlargest(k,'SalePrice')['SalePrice']
cols
#nlargest k number which column, return which column
```

```python
cols=cols.index
```

```python
plt.figure(figsize=(8,10))
sns.heatmap(house_data[cols].corr(),annot=True,cmap='coolwarm')
plt.show()

#only positive correlation here. in machine learning should take absolute functions!!!!
#coz negative correlation also important
```

```python
cols=['SalePrice', 'OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea']
sns.pairplot(house_data,vars=cols)
plt.show()
```

```python
cols=['SalePrice', 'OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea']
sns.pairplot(house_data,vars=cols,hue='SaleType')
plt.show()
#blue WD is more, warranty deed
```

# plot multiple chart together


```python
plt.figure(figsize=(12,8))
#space between charts
plt.subplots_adjust(hspace=0.3,wspace=0.3)

plt.subplot(2,2,1) #row num, col num, position
sns.kdeplot(x=house_data['SalePrice'])
plt.title('Price shart')

plt.subplot(2,2,2)
sns.histplot(x='SalePrice',data=house_data,hue='SaleCondition')
plt.title('Price hist')

plt.subplot(2,2,3)
sns.boxplot(x=house_data['SalePrice'])
plt.title('Boxplot for price')

plt.subplot(2,2,4)
plt.hist(house_data['SalePrice'],bins=50,color='g',histtype='step')
plt.title('hist for price')
plt.tight_layout()
plt.show()
```
