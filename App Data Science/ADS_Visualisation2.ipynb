```python
import pandas as pd
import matplotlib.pyplot as plt

url1 = 'https://raw.githubusercontent.com/shawnkan/Simpli/main/ADANIPORTS.csv'
```

```python
df = pd.read_csv(url1, parse_dates=True)
```

## Step 2: Add New Columns to the DataFrame

- Calculate the difference between the 'High' and 'Low' columns, and store it in the 'H-L' column
- Calculate the 100-day moving average of the 'Close' column, and store it in the '100MA' column


```python
df['H-L'] = df.High - df.Low
df['100MA'] = df['Close'].rolling(100).mean()
```

```python
df.shape
```

3D Scatter Plot 1

- Create a 3D scatter plot with the DataFrame index on the x-axis, 'H-L' on the y-axis and '100MA' on the z-axis
- Set x-axis, y-axis and z-axis labels
- Display the plot


```python
ax = plt.axes(projection='3d')
ax.scatter(df.index,df['H-L'],df['100MA'])
ax.set_xlabel('Index')
ax.set_ylabel('H-L')
ax.set_zlabel('100MA')
plt.show()
```

```python
ax = plt.axes(projection='3d')
ax.scatter(df.index,df['H-L'],df['Volume'])
ax.set_xlabel('Index')
ax.set_ylabel('H-L')
ax.set_zlabel('Volume')
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Create random data
np.random.seed(42)
n_points = 100
x = np.random.rand(n_points)
y = np.random.rand(n_points)
z = np.random.rand(n_points)

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='r', marker='o')

# Set labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Set title
ax.set_title('3D Scatter Plot Example')

plt.show()

#save for future
plt.savefig('Scatter3d.jpg')
```

linspace in numpy is used to create equal interval charts, equal data points

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Create data
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

# Create a 3D line plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, label='3D Spiral Line')

# Set labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Create data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(x, y, z, cmap='viridis')

# Add a color bar which maps values to colors
fig.colorbar(surf)

# Set labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Set title
ax.set_title('3D Surface Plot Example')

plt.show()
```
