```python
#import the lib

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
```

```python
#load the dataset
data =datasets.fetch_california_housing()
```

```python
print(data.feature_names)
```

```python
print(data.data)
```

```python
df = pd.DataFrame(data.data)
```

```python
df.columns = data.feature_names
df_target = np.asarray(data.target)
df['House_price']=pd.Series(df_target)
```

```python
df
```

```python
#extract dep & indep var
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test =train_test_split(X,y,test_size=.25, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#apply linear regression
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
#train
regressor.fit(X_train, y_train)
```

```python
#coefficient
print('Reg coeff: ', regressor.coef_)
print('y intercept', regressor.intercept_)
```

```python
y_pred = regressor.predict(X_test)
```

```python
#calculate the error
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
print('R2 score',r2_score(y_test,y_pred))
```

```python
X_train_pred = regressor.predict(X_train)
```

```python
print('R2 score',r2_score(y_train,X_train_pred))
```

```python
#usually can don't put alpha?
#apply ridge
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=10)
```

```python
#help(ridge_reg)
```

```python
#train
ridge_reg.fit(X_train, y_train)
```

```python
y_pred_ridge = ridge_reg.predict(X_test)
```

```python
print('R2 score',r2_score(y_test,y_pred_ridge))
```

```python
X_train_ridge = ridge_reg.predict(X_train)
```

```python
print('R2 score',r2_score(y_train,X_train_ridge))
```

```python
print(ridge_reg.coef_)
```

```python
ridge_coef=pd.DataFrame()
ridge_coef['Columns']=X_train.columns
ridge_coef['coef']=pd.Series(ridge_reg.coef_)
ridge_coef
```

```python

#apply lasso
from sklearn.linear_model import Lasso
lasso_reg = Lasso()
```

```python
#train
lasso_reg.fit(X_train, y_train)
```

```python
y_pred_lasso = lasso_reg.predict(X_test)
```

```python
print('R2 score',r2_score(y_test,y_pred_lasso))
```

```python
X_train_lasso = lasso_reg.predict(X_train)
```

```python
print('R2 score',r2_score(y_train,X_train_lasso))
```

```python
lasso_coef = pd.DataFrame()
lasso_coef['Columns']=X_train.columns
lasso_coef['coef']=pd.Series(lasso_reg.coef_)
lasso_coef
```
