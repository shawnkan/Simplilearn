```python
#import the lib

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load the dataset
data = pd.read_csv('https://raw.githubusercontent.com/shawnkan/Simplilearn/main/ML/Salary_Data.csv')
print('Data Loaded')
```

```python
data
```

```python
data.shape
```

```python
data.describe()
```

```python
plt.scatter(x=data['YearsExperience'], y=data['Salary'])
plt.show()
```

```python
#check missing values
data.isnull().sum()
```

```python
# Machine Learning
"""
1. Seperate indep and dep variable
2. split the training and testing set
3. Check for categorical col -> convert into num
4. scaling/normalisation
5. apply ML Algo
6. Evaluate the results
7. prediction
"""
```

```python
#Step 1
X= data.iloc[:,:-1].values
y= data.iloc[:,-1].values
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test =train_test_split(X,y,test_size=.2, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
"""
random state -
1 used to initialise random number gen
2. control the randomness
3. set a random state -ensure reproducibility - get the same results every time run the algo
"""
```

```python
#apply linear regression
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
#train
regressor.fit(X_train, y_train)
```

```python
#coefficient
print('Reg coeff: ', regressor.coef_)
print('y intercept', regressor.intercept_)
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
#calculate the error
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
print('MAE',mean_absolute_error(y_test,y_pred))
```

```python
print('MSE', mean_squared_error(y_test,y_pred))
```

```python
print('RMSE', np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
# help(regressor)
```

```python
# save the model - pickle, joblib
import joblib
joblib.dump(regressor,'Salary_pred_model.pkl')
print('Model is saved')
```

```python
print('R2 score',r2_score(y_test,y_pred))
```

```python
plt.scatter(X_test,y_test, color='b')
plt.plot(X_test,y_pred, color='g')
plt.title('Sal vs exp')
plt.show()
```

```python
X_train_pred = regressor.predict(X_train)
```

```python
plt.scatter(X_train,y_train, color='b')
plt.plot(X_train,X_train_pred, color='g')
plt.title('Sal vs exp')
plt.show()
```

```python
print('R2 score',r2_score(y_train,X_train_pred))
```
