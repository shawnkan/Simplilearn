```python
# import lib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
print('all lib imported')
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/lucko515/regression-python/master/Salaries%20based%20on%20Positions%20-%20dataset/Position_Salaries.csv')
```

```python
data
```

```python
#fit the regression model
X = data.iloc[:,1:2].values
y=data.iloc[:,2].values
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=42)
```

```python
#apply linear regression
#not enough data just apply fit on full X and y
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X, y)
y_pred=lr.predict(X)
```

```python
plt.scatter(X,y,color='red')
plt.plot(X,y_pred,color='blue')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
